```python
df.drop(columns=["Car_Name","Owner"],inplace=True)
```

```python
df.head(5)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
col_list=("Fuel_Type","Seller_Type","Transmission")
for col in col_list:
    df[col]=le.fit_transform(df[col])
```

```python
df.head(5)
```

```python
df.dtypes
```

```python
df.columns
```

```python
df=df[['Year', 'Transmission', 'Present_Price', 'Kms_Driven', 'Fuel_Type',
       'Seller_Type', 'Selling_Price']]
```

```python
df.head(5)
```

### BUILDING MACHNE LEARNING MODEL

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

### SPLITTING DATA INTO TRAIN AND TEST

```python
X=df.iloc[:,:-1]
y=df["Selling_Price"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
from sklearn.ensemble import AdaBoostRegressor
model=AdaBoostRegressor()
```

### TRAINING DATA USING ADABOOST MODEL

```python
model.fit(X_train,y_train)
pred=model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(pred,y_test)*100
```
